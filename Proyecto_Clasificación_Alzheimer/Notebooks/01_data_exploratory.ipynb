## üéØ 9. Detecci√≥n de Valores At√≠picos (Outliers)

### ¬øQu√© son los Outliers?
Los **valores at√≠picos** son observaciones que se desv√≠an significativamente del patr√≥n general de los datos.

### M√©todo IQR (Rango Intercuart√≠lico):
- **Q1**: Primer cuartil (25% de los datos)
- **Q3**: Tercer cuartil (75% de los datos)
- **IQR**: Q3 - Q1 (rango intercuart√≠lico)
- **Outliers**: Valores < Q1 - 1.5√óIQR o > Q3 + 1.5√óIQR

### ¬øPor qu√© es importante detectarlos?
- Pueden **distorsionar las estad√≠sticas** descriptivas
- Pueden **afectar modelos** de machine learning
- Pueden indicar **errores de medici√≥n** o casos especiales
- En medicina, pueden representar **condiciones at√≠picas** importantes

```python
# Filtrar solo las columnas num√©ricas
numeric_df = df.select_dtypes(include=[np.number])

# Check outliers IQRS
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
outliers_iqr = ((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR)))
print("Outliers detected using IQR method:\n", outliers_iqr.sum())

# Mostrar los valores at√≠picos detectados por el m√©todo IQR
outliers_iqr_values = numeric_df[outliers_iqr.any(axis=1)]
print("Valores at√≠picos detectados usando el m√©todo IQR:\n", outliers_iqr_values)
```

## üîÑ 10. Transformaci√≥n de Variables

```python
# Sustituir la columna de edad por otra con rangos
df['Age'] = pd.cut(df['Age'], bins=[0, 65, 75, 85, 95, 100], labels=['0-65', '66-75', '76-85', '86-95', '96-100'])
```

```python
# Show unique values for gender column

print("Unique values for Gender column: ", df["Gender"].unique())
```

```python
# Sustituir 0 por Male y 1 por Female

df['Gender'] = df['Gender'].replace({0: 'Male', 1: 'Female'})
```

```python
# Substitute Ethnicity codes with corresponding names
df['Ethnicity'] = df['Ethnicity'].replace({0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'})
```

```python
# Substitute EducationLevel codes with corresponding names
df['EducationLevel'] = df['EducationLevel'].replace({0: 'No Education', 1: 'High School', 2: "Bachelor's", 3: 'Higher'})
```

```python
# Sustituir y hacer rangos para los factores de estilo de vida

# BMI
df['BMI'] = pd.cut(df['BMI'], bins=[15, 20, 25, 30, 35, 40], labels=['15-20', '21-25', '26-30', '31-35', '36-40'])

# Smoking (0 indica no, 1 indica si)
df['Smoking'] = df['Smoking'].replace({0: 'No', 1: 'Yes'})

# AlcoholConsumption
df['AlcoholConsumption'] = pd.cut(df['AlcoholConsumption'], bins=[0, 5, 10, 15, 20], labels=['0-5', '6-10', '11-15', '16-20'])

# PhysicalActivity
df['PhysicalActivity'] = pd.cut(df['PhysicalActivity'], bins=[0, 2, 4, 6, 8, 10], labels=['0-2', '3-4', '5-6', '7-8', '9-10'])

# DietQuality
df['DietQuality'] = pd.cut(df['DietQuality'], bins=[0, 2, 4, 6, 8, 10], labels=['0-2', '3-4', '5-6', '7-8', '9-10'])

# SleepQuality
df['SleepQuality'] = pd.cut(df['SleepQuality'], bins=[4, 5, 6, 7, 8, 9, 10], labels=['4-5', '6', '7', '8', '9', '10'])
```

```python
# Sustituir y hacer rangos para el historial m√©dico

# FamilyHistoryAlzheimers
df['FamilyHistoryAlzheimers'] = df['FamilyHistoryAlzheimers'].replace({0: 'No', 1: 'Yes'})

# CardiovascularDisease
df['CardiovascularDisease'] = df['CardiovascularDisease'].replace({0: 'No', 1: 'Yes'})

# Diabetes
df['Diabetes'] = df['Diabetes'].replace({0: 'No', 1: 'Yes'})

# Depression
df['Depression'] = df['Depression'].replace({0: 'No', 1: 'Yes'})

# HeadInjury
df['HeadInjury'] = df['HeadInjury'].replace({0: 'No', 1: 'Yes'})

# Hypertension
df['Hypertension'] = df['Hypertension'].replace({0: 'No', 1: 'Yes'})
```

```python
# Obtener valores √∫nicos de SystolicBP y DiastolicBP
unique_systolic_bp = df['SystolicBP'].unique()
unique_diastolic_bp = df['DiastolicBP'].unique()

print("Unique values for SystolicBP column: ", unique_systolic_bp)
print("Unique values for DiastolicBP column: ", unique_diastolic_bp)
```

```python
# Sustituir y hacer rangos para las mediciones cl√≠nicas

# SystolicBP
df['SystolicBP'] = pd.cut(df['SystolicBP'], bins=[90, 110, 130, 150, 170, 180], labels=['90-110', '111-130', '131-150', '151-170', '171-180'])

# DiastolicBP
df['DiastolicBP'] = pd.cut(df['DiastolicBP'], bins=[60, 70, 80, 90, 100, 120], labels=['60-70', '71-80', '81-90', '91-100', '101-120'])

# CholesterolTotal
df['CholesterolTotal'] = pd.cut(df['CholesterolTotal'], bins=[150, 180, 210, 240, 270, 300], labels=['150-180', '181-210', '211-240', '241-270', '271-300'])

# CholesterolLDL
df['CholesterolLDL'] = pd.cut(df['CholesterolLDL'], bins=[50, 90, 130, 170, 200], labels=['50-90', '91-130', '131-170', '171-200'])

# CholesterolHDL
df['CholesterolHDL'] = pd.cut(df['CholesterolHDL'], bins=[20, 40, 60, 80, 100], labels=['20-40', '41-60', '61-80', '81-100'])

# CholesterolTriglycerides
df['CholesterolTriglycerides'] = pd.cut(df['CholesterolTriglycerides'], bins=[50, 150, 250, 350, 400], labels=['50-150', '151-250', '251-350', '351-400'])
```

```python
# Sustituir y hacer rangos para las evaluaciones cognitivas y funcionales

# MMSE
df['MMSE'] = pd.cut(df['MMSE'], bins=[0, 10, 20, 30], labels=['0-10', '11-20', '21-30'])

# FunctionalAssessment
df['FunctionalAssessment'] = pd.cut(df['FunctionalAssessment'], bins=[0, 2, 4, 6, 8, 10], labels=['0-2', '3-4', '5-6', '7-8', '9-10'])

# MemoryComplaints
df['MemoryComplaints'] = df['MemoryComplaints'].replace({0: 'No', 1: 'Yes'})

# BehavioralProblems
df['BehavioralProblems'] = df['BehavioralProblems'].replace({0: 'No', 1: 'Yes'})

# ADL
df['ADL'] = pd.cut(df['ADL'], bins=[0, 2, 4, 6, 8, 10], labels=['0-2', '3-4', '5-6', '7-8', '9-10'])
```

```python
# Sustituir y hacer rangos para los s√≠ntomas

# Confusion
df['Confusion'] = df['Confusion'].replace({0: 'No', 1: 'Yes'})

# Disorientation
df['Disorientation'] = df['Disorientation'].replace({0: 'No', 1: 'Yes'})

# PersonalityChanges
df['PersonalityChanges'] = df['PersonalityChanges'].replace({0: 'No', 1: 'Yes'})

# DifficultyCompletingTasks
df['DifficultyCompletingTasks'] = df['DifficultyCompletingTasks'].replace({0: 'No', 1: 'Yes'})

# Forgetfulness
df['Forgetfulness'] = df['Forgetfulness'].replace({0: 'No', 1: 'Yes'})

# Diagnosis
df['Diagnosis'] = df['Diagnosis'].replace({0: 'No', 1: 'Yes'})
```
